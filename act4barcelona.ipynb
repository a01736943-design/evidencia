```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .csv sin indice
df= pd.read_csv('listingsbarcelona.csv')
df
```

```python
df.info()
```

## Los registros de 5 “host name” diferentes
## Los registros de los host que se hallan unido a Airbnb “despues del año 2020”
## Los registros de los host que responden “a más tardar en 1 día”

```python
top5_hosts = df["host_name"].value_counts().head(5).index.tolist()
filtro1 = df[df["host_name"].isin(top5_hosts)]
filtro1.head()
```

```python
filtro2 = df[df["host_since"] > "2020-12-31"]
filtro2
```

```python
filtro3 = df[df["host_response_time"] == "within a day"]
filtro3
```

### os registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”
### Los registros de los tipos de cuarto: “Entire home” y “Private room”
### Los registros que “a lo mucho cuenten con 4 camas

```python
filtro4 = df[df.index % 200 == 0]
filtro4
```

```python
filtro5 = df[df["room_type"].isin(["Entire home", "Private room"])]
filtro5
```

```python
filtro6 = df[df["beds"] <= 4]
filtro6
```

## Los registros que “sean superhost”
## Los registros de las columnas “impares”

```python
filtro7 = df[df["host_is_superhost"] == "t"]
filtro7
```

```python
filtro8 = df.iloc[:, 1::2]
filtro8
```

```python
#Convertir archivos filtrados a CSV
filtro1.to_csv("filtro1barcelona.csv")
filtro2.to_csv("filtro2barcelona.csv")
filtro3.to_csv("filtro3barcelona.csv")
filtro4.to_csv("filtro4barcelona.csv")
filtro5.to_csv("filtro5barcelona.csv")
filtro6.to_csv("filtro6barcelona.csv")
filtro7.to_csv("filtro7barcelona.csv")
filtro8.to_csv("filtro8barcelona.csv")
```
